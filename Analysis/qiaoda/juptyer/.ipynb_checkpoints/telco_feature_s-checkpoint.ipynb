```python
def get_sex(x):
    if isinstance(x, unicode):
        if u"男" in x:
            return u'男'
        elif u"女" in x:
            return u"女"
        else:
            return u"未知"
    else:
        return u"未知"
```

```python
r_age = re.compile(ur"年龄段.*?([0-9]+-?[0-9]*)")
def get_ages(x):
    try: 
        ages =  r_age.findall(x)[0].split("-")
        ages_len = len(ages)
        if ages_len == 0:
            return [None, None]
        elif ages_len == 1:
            return [int(ages[0]), int(ages[0])]
        else:
            return [int(ages[0]), int(ages[1])]
    except:
        return [None, None]
```

```python
df_t[u'年龄下限'], df_t[u'年龄上限'] = zip(*df_t[u'基本特征'].apply(lambda x: get_ages(x)))
```

```python
re_edu = re.compile(ur';(本科.*?|研究生.*?|博士.*?|专科.*?|高中.*?|小学.*?|初中.*?|中专.*?);')
def get_edu(x):
    try:
        edu = re_edu.findall(x)[0]
        return edu
    except:
        return u"未知"
```

提取经济水平特征：

- 按照特征中是否存在： 有车， 有房， 较高等关键字， 如果有视为高收入人群， 给予1

```python
re_eco = re.compile(ur'(\b有车\b|\b有房\b|\b较高\b)', re.UNICODE)
def get_eco_level(x):
    try:
        s = re_eco.search(x)
        if x == u'未知':
            return 2
        elif s:
            return 1
        else:
            return 0
    except:
        return 0
```

```python
df_t[u'经济水平'] = df_t[u'经济水平'].astype(unicode)
```

```python
df_t[u'经济水平'] = df_t[u'经济水平'].apply(lambda x: get_eco_level(x))
```

提取消费水平特征：

- 根据特征中是否存在团购，购物，支付关键字， 存在给予1

```python
re_comsu = re.compile(ur'(\b购物\b|\b团购\b|\b支付\b)', re.UNICODE)
def get_consume_feature(x):
    try:
        s = re_comsu.search(x)
        if x == u'未知':
            return 2
        elif s:
            return 1
        else:
            return 0
    except:
        return 0
```

```python
df_t[u'消费特征'] = df_t[u'消费特征'].apply(lambda x: get_consume_feature(x))
```

提取行为偏好特征：

```python
re_act = re.compile(ur'(借款|欠款|借贷|还款|欠款|欠贷|借|网贷)', re.UNICODE)
def get_act_feature(x):
    try:
        s = re_act.search(x)
        if x == u'未知':
            return 2
        elif s:
            return 1
        else:
            return 0
    except:
        return 0
```

提取个人信用特征：

- 是否经常换卡
- 是否使用小贷APP

```python
re_new_card = re.compile(ur'经常换卡', re.UNICODE)
re_app = re.compile(ur'小贷app总个数.*?([0-9,\|]+)', re.UNICODE)
def get_credit(x):
    try:
        if x == u'未知':
            new_card = 2
        elif re_new_card.search(x):
            new_card = 1
        else:
            new_card = 0
        
        if x == u'未知':
            app = 2
        elif re_app.search(x):
            app = 1
        else:
            app = 0
        return (new_card, app)
    except:
        return(0,0)
```

```python
chi_cols = df_mt.columns.difference(['IMEI', u'年龄下限', u'年龄上限','label','fraud_type', 'telco'])
```

```python
def chi2_state(df_mt, df_t,chi_cols,filter_value=None):
    chi2_p = []
#     decode_cols = {}
    if filter_value:
        for col in chi_cols:
            print("去除无效值", col)
            df_cm = pd.DataFrame({'encode':df_mt[col], 'basic':df_t[col]}).groupby(['basic','encode']).agg('count').reset_index()
            if df_cm.basic.dtype == unicode or df_cm.basic.dtype == str or df_cm.basic.dtype == np.object_:
                non_code = df_cm.loc[df_cm.basic == u'未知']['encode'].values[0]
                dec_cols = {col:dict(zip(df_cm.encode, df_cm.basic))}
            else:
                non_code = 2
                dec_cols = {col:{1:u'是',2:u'未知', 0:u'否'}}
            df_mt_new = df_mt.loc[df_mt[col] != non_code]
            cross_table = pd.crosstab(df_mt_new.label, df_mt_new[col])
            print("chi2 state result:")
            p = chi2_contingency(cross_table)[1]
            print(p)
            chi2_p.append([p,non_code,dec_cols])
            print("end----")
        return chi2_p
    else:
        for col in chi_cols:
            print("不去除无效值",col)
            df_cm = pd.DataFrame({'encode':df_mt[col], 'basic':df_t[col]}).groupby(['basic','encode']).agg('count').reset_index()
            if df_cm.basic.dtype == unicode or df_cm.basic.dtype == str or df_cm.basic.dtype == np.object_:
                non_code = df_cm.loc[df_cm.basic == u'未知']['encode'].values[0]
                dec_cols = {col:dict(zip(df_cm.encode, df_cm.basic))}
            else:
                non_code = 2
                dec_cols = {col:{1:u'是',2:u'未知', 0:u'否'}}
            cross_table = pd.crosstab(df_mt.label, df_mt[col])
            print("chi2 state result:")
            p = chi2_contingency(cross_table)[1]
            print(p)
            chi2_p.append([p,non_code,dec_cols])
            print("end----")
        return chi2_p
```

```python
chi2_p=[]
dec_cols=[]
for x in chi2_state(df_mt=df_mt, df_t=df_t, chi_cols=chi_cols):
    chi2_p.append(x[0])
    dec_cols.append(x[2])
```

```python
def pic_chi2(chi2_p, chi2_cols):
    fig,ax= plt.subplots(figsize=(10,6))
    chi2_df = pd.DataFrame(chi2_p, index=chi_cols, columns=['chi2_p'])
    chi2_log = chi2_df.plot.bar(ax=ax)
    log_bin_middles = np.array(map(lambda patch:patch._x + patch._width / 2.0 , chi2_log.containers[0]))
    ax.plot(log_bin_middles,[0.05] * len(chi2_p), c='r')
    ax.plot(log_bin_middles,[0.1] * len(chi2_p), c='g')
    plt.show()
```
